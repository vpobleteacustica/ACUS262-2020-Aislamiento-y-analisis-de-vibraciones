> + En una representación más compacta:

$$
\begin{align}
\text{M}{\ddot{{\bf{x}}}} + \text{K}{\bf{x}} &={\bf{F}}
\end{align}
$$

> + donde $\text{M}$ se llama **matriz de masa**, $\text{K}$ se denomina **matriz de rigidez**, $\bf{\text{F}}$ es el **vector de fuerzas**, y ${\bf{x}}$ es llamado **vector que contiene los grados de libertad del sistema**.

> + En este ejemplo, ${\bf{x}}$ contiene a $x_{1}$ y $x_{2}$. 

> + La forma de la representación compacta se puede generalizar para escribir las ecuaciones de movimiento de un sistema de $n$ grados de libertad.

> + En ese caso, el tamaño de las matrices sería de $n\times n$ y el tamaño de los vectores sería $n$.

### La matriz de modos:

> + La importancia de esta matriz, es que se usa para definir una **transformación de coordenadas.** 

> + Para ésto, se introduce una nueva coordenada ${\bf{y}}$, a veces llamada **Coordenada Principal**.

> + La transformación de coordenadas es ${\bf{x}}(t)=P{\bf{y}}(t).$

###  La solución general de la ecuación de movimiento se escribe como la sumatoria de los productos entre: 
>> + Los vectores forma de vibrar o *mode shapes*, en el **dominio espacial**, 
>> + los factores $d_{i}$ que indican cuánto aporta cada modo en la respuesta total y que dependen de las condiciones iniciales: deplazamiento y velocidad inicial.


$$
\begin{align*}
d_{i} & = \frac{\sqrt{\omega_{i}^{2}y_{i0}^{2} + \dot{y}_{i0}^{2} }}{\omega_{i}}
\end{align*}
$$

>> + Las funciones base $\sin()$ o $\cos()$ oscilatorias en el dominio del tiempo; oscilaciones que se encuentran en las frecuencias $\omega_{i}$.
$$
\begin{align*}
x(t) & = P\,y(t)\\
     & = \sum_{i=1}^{n} {\bf{u}}_{i}y_{i}(t)\\
     & = \sum_{i=1}^{n} {\bf{u}}_{i}d_{i}\sin\left( \omega_{i}t + \phi_{i}\right)
\end{align*}
$$

###  Dos condiciones de ortogonalidad de los vectores forma de vibrar (mode shapes):

> + Es muy útil explicarlas usando la siguiente notación de delta de Kronecker:

> + Consideremos dos valores propios $\omega_{i}$ y $\omega_{j}$, y dos vectores propios ${\bf{u}}_{i}$ y ${\bf{u}}_{j}$.

> + **Primera condición**:
$$
\begin{align*}
{\bf{u}}_{i}^{T}\text{M}\,{\bf{u}}_{j} & = \delta_{ij}\cdot m_{ij} = \left\{ \begin{array}{cc} 
               m_{ii}, & {\text{si}}\,\, i=j \\
               0,  & {\text{si}}\,\, i\neq j  
                \end{array}\right.
\end{align*}
$$

> + **Segunda condición**:

$$
\begin{align*}
{\bf{u}}_{i}^{T}\text{K}\,{\bf{u}}_{j} & = \delta_{ij}\cdot k_{ij} = \delta_{ij}\cdot m_{ij}\omega_{i}^{2} = \left\{ \begin{array}{cc} 
               m_{ii}\cdot\omega_{i}^{2}, & {\text{si}}\,\, i=j \\
               0,  & {\text{si}}\,\, i\neq j  
                \end{array}\right.
\end{align*}
$$

> +  La masa $m_{ii}$ se llama masa generalizada o  también se le llama **masa del modo**.

> +  La rigidez $k_{ii}=  m_{ij}\omega_{i}^{2}$ se llama rigidez generalizada o  también se le llama **rigidez del modo**.

> + La frecuencia $\omega_{i} = 2\pi f_{i}$ es la frecuencia natural del modo $i-$ésimo.

> + Estas dos condiciones de ortogonalidad se suelen escribir en forma matricial. Para explicarlo, comencemos escribiendo la matriz de modos $\left[P\right]$ cuyas columnas son los vectores modos de vibrar (*mode shapes*)

$$
\begin{align*}
P & = \begin{bmatrix}
 {\bf{u}}_{1} & {\bf{u}}_{2} & \ldots & {\bf{u}}_{n}
\end{bmatrix}
\end{align*}
$$

> + Y obtenemos las condiciones de ortogonalidad en representación matricial:

$$
\begin{align*}
P^{T}M\,P & = \begin{bmatrix}
m_{11} & 0 & \ldots & 0\\
0      & m_{22} & \ldots & 0\\
\vdots & \vdots & \vdots & \vdots\\
0      & 0      & \ldots & m_{nn}\\
\end{bmatrix}=\text{diag}\left( m_{ii}\right)
\end{align*}
$$

$$
\begin{align*}
P^{T}K\,P & = \begin{bmatrix}
m_{11}\omega_{1}^{2} & 0 & \ldots & 0\\
0      & m_{22}\omega_{2}^{2} & \ldots & 0\\
\vdots & \vdots & \vdots & \vdots\\
0      & 0      & \ldots & m_{nn}\omega_{n}^{2}\\
\end{bmatrix}=\text{diag}\left( m_{ii}\omega_{i}^{2}\right)
\end{align*}
$$

> + Lo más importante de estas dos condiciones matriciales de ortogonalidad es que expresan que los vectores de modos de vibrar pueden usarse como una **base ortogonal** que permite representar la solución del problema. 

### Solución particular de la ecuación de movimiento:

> + Comencemos con la ecuación compacta:

$$
\begin{align}
\text{M}{\ddot{{\bf{x}}}} + \text{K}{\bf{x}} &={\bf{F}}
\end{align}
$$

> + Usemos los vectores de modos de vibrar como una **base ortogonal** para representar la solución:

$$
\begin{align}
{\bf{x}}(t)=P{\bf{y}}(t)
\end{align}
$$

> + donde ${\bf{y}}$ es la coordenada **principal** que representa las amplitudes del modo.

> + Reemplacemos ${\bf{x}}$ por $P{\bf{y}}$ en ec. (1) y pre-multipliquemos por $P^{T}$:

$$
\begin{align*}
P^{T}\text{M}{P\,\ddot{{\bf{y}}}} + P^{T}\text{K}P\,{\bf{y}} & = P^{T}{\bf{F}}
\end{align*}
$$

> + Usamos ahora las dos condiciones de ortogonalidad anteriormente explicadas:

$$
\begin{align}
\text{diag}\left( m_{ii}\right)\,\ddot{{\bf{y}}}+ \text{diag}\left( m_{ii}\omega_{i}^{2}\right){\bf{y}} & = P^{T}{\bf{F}}
\end{align}
$$

> + Esto corresponde a un conjunto de $n$ **ecuaciones desacopladas** del tipo:

$$
\begin{align}
m_{ii}\,\ddot{y} + m_{ii}\omega_{i}^{2}y & = F_{i}
\end{align}
$$

> + La ec. (3) es una ecuación de movimiento de un típico sistema de un grado de libertad con:
>> + una masa $m_{ii}$ llamada **masa del modo**, 
>> + una rigidez $m_{ii}\omega_{i}^{2}$ denominada **rigidez del modo**,
>> + una **frecuencia natural** $\omega_{i}=2\pi f_{i}$, 
>> + y una fuerza $F_{i}=u_{i}^{T}f$ llamada **excitación del modo**.

### En resumen:

> + Al escribir la solución particular ${\bf{x}}(t)$ como una función de $y_{i}(t)$ y los vectores modales $u_{i}$, se transforma el conjunto inicial de $n$ ecuaciones de movimiento **acopladas** en un conjunto de $n$ ecuaciones de movimiento **desacopladas**.

> + Cada ecuación de movimiento desacoplada representa la ecuación independiente de un típico **sistema de un grado de libertad**. 

> + Todos los recursos descritos hasta ahora se pueden usar para representar la solución de las ecuaciones de movimiento del sistema de múltiples grados de libertad.

### Solución particular en el dominio de la frecuencia:

> + Supongamos una excitación de tipo armónica.

> + Aplicamos transformada de Fourier a:

$$
\begin{align}
\text{M}{\ddot{{\bf{x}}}} + \text{K}{\bf{x}} &={\bf{F}}
\end{align}
$$

> + y ordenamos términos:

$$
\begin{align}
\left( \text{K} - \omega^{2}\text{M}\right)X &=F
\end{align}
$$

> + donde $X$ y $F$ son vectores de tamaño $n$.

> + Para resolver la Ec. (2), usemos los vectores de modos de vibrar como una **base ortogonal** para representar la solución $X$:

$$
\begin{align*}
X(\omega) & = \sum_{i=1}^{n}Y_{i}(\omega)\,u_{i}
\end{align*}
$$

> + Y la escribimos en forma matricial:

$$
\begin{align*}
X & = P Y
\end{align*}
$$

> + donde $Y$ es un vector de amplitudes de los modos.

> + Sustituyamos en la Ec. (2) y pre-multipliquemos por $P^{T}$:

$$
\begin{align}
\left( {\underbrace{P^{T}\text{K}P}_{ \text{diag}\left( m_{ii}\omega_{i}^{2}\right) }} - \omega^{2}{\underbrace{P^{T}\text{M}P}_{ \text{diag}\left( m_{ii}\right) }} \right)\begin{bmatrix}
 Y_{1}\\
 Y_{2} \\
 \vdots \\
 Y_{n}
\end{bmatrix} & = \begin{bmatrix}
 u_{1}^T F\\
 u_{2}^T F \\
 \vdots \\
 u_{n}^T F
\end{bmatrix}
\end{align}
$$

> + Obtenemos un conjunto de $n$ ecuaciones desacopladas.

> + Las incógnitas $Y_{i}$ se re-organizan como:

$$
\begin{align*}
Y_{i}(\omega) & = u_{i}^T F\,\frac{1}{m_{ii}\left( \omega_{j}^{2} - \omega^{2}\right)}
\end{align*}
$$

> + El desplazamiento total $X(\omega)$ será la sumatoria d elas contribuciones de los diferentes modos de vibrar:

$$
\begin{align*}
X(\omega) & = \sum_{i=1}^{n} Y_{i}(\omega)\,u_{i}\\
          & = \sum_{i=1}^{n} \frac{u_{i}^T F}{m_{ii}\left( \omega_{j}^{2} - \omega^{2}\right)}
\end{align*}
$$

> + El diagrama de $X(\omega)$ presentará un máximo de $n$ peaks resonantes.

### Volvamos a nuestro ejemplo:

> + La matriz de modos $\left[P\right]$ cuyas columnas son los vectores modos de vibrar (*mode shapes*) es:

$$
\begin{align*}
P & = \begin{bmatrix}
 u_{1}& u_{2}
\end{bmatrix} =\frac{\sqrt{2}}{2}\begin{bmatrix}
 1 & 1\\
 1 & -1
\end{bmatrix}
\end{align*}
$$

> + Calculamos los valores de $m_{ii}$:

$$
\begin{align*}
P^{T}M\,P & = \frac{\sqrt{2}}{2}\,\begin{bmatrix}
 1 & 1\\
 1 & -1
\end{bmatrix}\,\begin{bmatrix}
m & 0 \\
0 & m
\end{bmatrix}\cdot\,\frac{\sqrt{2}}{2}\begin{bmatrix}
 1 & 1\\
 1 & -1
\end{bmatrix}\\
& = \begin{bmatrix}
 m & 0\\
 0 & m
\end{bmatrix}
\end{align*}
$$

> + Esto es: $m_{11}=m_{22}=m$.

> + Los valores de $F_{i} = P^{T}F$ (ver Ecs. 3 y 4) son:

$$
\begin{align*}
P^{T}F & = \frac{\sqrt{2}}{2}\,\begin{bmatrix}
 1 & 1\\
 1 & -1
\end{bmatrix}\,\begin{bmatrix}
0 \\
F
\end{bmatrix}= \frac{\sqrt{2}}{2}\begin{bmatrix}
 F(\omega)\\
 -F(\omega)
\end{bmatrix}\\
& = \frac{\sqrt{2}}{2}F(\omega)\begin{bmatrix}
 1\\
 -1
\end{bmatrix}
\end{align*}
$$

> + Por lo tanto, las incógnitas $Y_{1}$ e $Y_{2}$ que representan las amplitudes de los modos son:

$$
\begin{align*}
Y_{1}(\omega) & = \frac{\sqrt{2}}{2}\frac{F(\omega)}{m\left( \omega_{1}^{2}-\omega^{2}\right)}\\
& = \frac{\sqrt{2}}{2m}\frac{F(\omega)}{\left( \frac{k}{m}-\omega^{2}\right)}
\end{align*}
$$

$$
\begin{align*}
Y_{2}(\omega) & = \frac{-\sqrt{2}}{2}\frac{F(\omega)}{m\left( \omega_{1}^{2}-\omega^{2}\right)}\\
& = -\frac{\sqrt{2}}{2m}\frac{F(\omega)}{\left( \frac{3k}{m}-\omega^{2}\right)}
\end{align*}
$$

> + Y el desplazamiento total $X(\omega)$ del sistema es:

$$
\begin{align*}
X(\omega) & = \begin{bmatrix}
 X_{1}(\omega) \\
 X_{2}(\omega)
\end{bmatrix}\\
& = Y_{1}(\omega)\cdot u_{1} + Y_{2}(\omega)\cdot u_{2}\\
& = \frac{\sqrt{2}}{2m}\frac{F(\omega)}{\left( \frac{k}{m}-\omega^{2}\right)}\cdot \frac{\sqrt{2}}{2}\begin{bmatrix}
 1 \\
 1 
\end{bmatrix} -\frac{\sqrt{2}}{2m}\frac{F(\omega)}{\left( \frac{3k}{m}-\omega^{2}\right)}\frac{\sqrt{2}}{2}\begin{bmatrix}
 1 \\
 -1 
\end{bmatrix}\\
& = \frac{F(\omega)}{2m\left( \frac{k}{m}-\omega^{2}\right)}\cdot \begin{bmatrix}
 1 \\
 1 
\end{bmatrix} - \frac{F(\omega)}{2m\left( \frac{3k}{m}-\omega^{2}\right)}\cdot \begin{bmatrix}
 1 \\
 -1 
\end{bmatrix}
\end{align*}
$$

### Importo librerías:

```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt

from matplotlib.ticker import ScalarFormatter
from matplotlib.ticker import EngFormatter
from scipy.signal import find_peaks
from matplotlib.offsetbox import AnchoredText
%matplotlib notebook
```

```python
w    = np.linspace(0,100,10000)

# Definimos los valores para los parámetros:
m   = 1.0
m_1 = m
m_2 = m

k   = 4.0
k_1 = k
k_2 = k
k_3 = k

# Los valores propios resultantes son: 4.00 and 12.00.
w1 = np.sqrt(k/m)
w2 = np.sqrt(3*k/m)

print(w1**2)
print(w2**2)
```

### Modo 1:

```python
XF1 = (1/((2*m)*(w1**2 - w**2))) - (1/((2*m)*(w2**2 - w**2)))
XF11 = (1/((2*m)*(w1**2 - w**2)))
XF12 = - 1/((2*m)*(w2**2 - w**2))
MagXF1  = np.abs(XF1)
MagXF11 = np.abs(XF11)
MagXF12 = np.abs(XF12)

plt.figure(figsize=(8, 4))
plt.plot(w/2*np.pi, 10*np.log(MagXF1**2))
plt.plot(w/2*np.pi, 10*np.log(MagXF11**2), linestyle='dashed')
plt.plot(w/2*np.pi, 10*np.log(MagXF12**2), linestyle='dashed')          

plt.grid()
plt.xlabel('frecuencia (Hz)')
plt.ylabel('amplitud')
plt.ylim(-160.0, 160.0)
plt.xlim(0.0, 10.0)
plt.title("Contribución de los modos en fase") 
```

### Modo 2:

```python
XF2 = (1/((2*m)*(w1**2 - w**2))) + (1/((2*m)*(w2**2 - w**2)))

XF21 = (1/((2*m)*(w1**2 - w**2)))
XF22 = 1/((2*m)*(w2**2 - w**2))

MagXF2 = np.abs(XF2)
MagXF21 = np.abs(XF21)
MagXF22 = np.abs(XF22)


plt.figure(figsize=(8, 4))
plt.plot(w/2*np.pi, 10*np.log(MagXF2**2))
plt.plot(w/2*np.pi, 10*np.log(MagXF21**2), linestyle='dashed')
plt.plot(w/2*np.pi, 10*np.log(MagXF22**2), linestyle='dashed')
plt.grid()
plt.xlabel('frecuencia (Hz)')
plt.ylabel('amplitud')
plt.xlim(0.0, 10.0)
plt.title("Contribución de los modos en desfase") 
```

### Para la masa $m_{1}$:

> + Las contribuciones de $Y_{1}(\omega)$ e $Y_{2}(\omega)$ están en fase, de modo que las amplitudes se suman.

> + Esto resulta en una amplitud más alta de $Y_{1}(\omega)$.

### Para la masa $m_{2}$:

> + Las contribuciones de $Y_{1}(\omega)$ e $Y_{2}(\omega)$ tienen fases opuestas.

> + En la frecuencia en la que sus amplitudes son iguales, la suma de las contribuciones es cero.

> + En esa frecuencia, la masa $m_{2}$ **no se mueve**.

> + La frecuencia se llama **antiresonancia**.

### Las frecuencias resonantes del sistema son:

> + Una propiedad **global** del sistema es decir, ellas están presentes para cada grado de libertad del sistema.

### Las antiresonancias son:

> + Una propiedad **local** del sistema, es decir, están presentes solamente en grados de libertad específicos del sistema.
