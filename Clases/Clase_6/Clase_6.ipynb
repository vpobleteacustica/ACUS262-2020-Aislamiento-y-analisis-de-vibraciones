#### Desarrollo

Usando la segunda ley de Newton y siguiendo las condiciones de borde mostradas de la Fig. 3, la ecuación de movimiento será:

\begin{align}
k_{2}\,(\,x_{1}-x\,\,)\,-k_{1}\,(\,x\,-\,x_{0}\,)\,&=\,m\,\ddot{x}(t)\\
%\label{eq2}\tag{2}
\end{align}

lo que se puede arreglar como:

\begin{align}
m\,\ddot{x}\,(t)\,+\,(k_{1}\,+\,k_{2}\,)\,x\,(t)\,&=\,0
\label{eq2}\tag{2}
\end{align}

de donde la constante de rigidez equivalente $k_{eq}$ del sistema está dada por:

\begin{align}
k_{eq}\,&=\,(\,k_{1}\,+\,k_{2}\,)
\label{eq3}\tag{3}
\end{align}

siendo $\omega_{0}$:

\begin{align}
\omega_{0}\,&=\,\sqrt{\frac{(\,k_{1}\,+\,k_{2}\,)}{m}}\quad\bigg(\frac{rad}{s}\bigg)
\label{eq4}\tag{4}
\end{align}

de modo que el período de vibración será:

\begin{align}
T&\,=\,\frac{2\pi}{\sqrt{\frac{(\,k_{1}\,+\,k_{2}\,)}{m}}}\,=\,2\pi\,\sqrt{\frac{m}{(\,k_{1}\,+\,k_{2}\,)}}\quad(s)
\label{eq5}\tag{5}
\end{align}

**1. Sistema mecánico rotatorio equivalente**
+ Comparando con el sistema de la Fig. 3 (a), se puede asumir el siguiente sistema mecánico rotatorio equivalente:

<p style="text-align:center">

+ La Fig. 5 modela una primera barra (eje de la izquierda) en vibración torsional, fija en el extremo izquierdo y sujeta a un torque $M$ en $x=L=1.6$ (m). Similarmente, la segunda barra (eje de la derecha) también en vibración torsional, fija en el extremo derecho y sujeta a un torque $M$ en $x=L=1.6$ (m). La rigidez equivalente $k_{eq}$ torsional de las barras estará dada por:

\begin{align}
k_{eq}&=\bigg(\,\frac{GJ_{1}}{L}\,+\,\frac{GJ_{2}}{L}\,\bigg)\quad\bigg(\frac{N\cdot m}{rad}\bigg)
%\label{eq12}\tag{12}
\end{align}

siendo $G$ el módulo de corte de las barras y $J_{1}$ y $J_{2}$, los momentos de inercia polares (o torsionales) de las barras ($m^{4}$). Para una barra cilíndrica sólida, de diámetro $d$, el momento de inercia polar se calcula como:

\begin{align}
J &=\frac{\pi}{32}\,d^{4}\quad(\,m^{4}\,)
\end{align}

* El desplazamiento torsional $\theta(t)$ va a satisfacer la ecuación diferencial de movimiento:

\begin{align}
I\,\ddot\theta(t)+k_{eq}\,\theta(t)=0
\end{align}

donde la masa del disco que rota está representada por $I$, llamado momento de inercia del disco. En este caso, el momento de inercia del disco cilíndrico de radio $r$, se calcula como:

\begin{align}
I&=\frac{1}{2}\,m\,r^{2}\quad(kg\cdot m^{2})
\end{align}

**2. Expresión general que define la oscilación armónica para el desplazamiento torsional** 

Para este caso, el sistema mecánico rotatorio equivalente, con sus dos condiciones iniciales, satisface la siguiente ecuación:

\begin{align}
I\,\ddot\theta(t)+k_{eq}\,\theta(t)=0; \quad \theta(0)=\theta_{0}, \quad \dot{\theta}(0)=\dot{\theta}_{0}=0
\label{eq2}\tag{2}
\end{align}

donde la solución general será:

\begin{align*}
\theta(t)&=c\,{\text{cos}}\,(\omega_{0}t-\phi)
\end{align*}

siendo $c$ igual a:

\begin{align*}
c&=\sqrt{\theta_{0}^{2}+\bigg( \frac{\dot{\theta}_{0}}{\omega_{0}}\bigg)^{2}}
\end{align*}

y $\phi$ se obtiene de:

\begin{align*}
\phi &= tg^{-1}\bigg(\frac{\dot{\theta}_{0}}{\theta_{0}\,\omega_{0}}\bigg)
\end{align*}
