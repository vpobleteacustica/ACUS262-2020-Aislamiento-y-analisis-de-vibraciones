### Sistemas vibratorios continuos y discretos:

> + Ciertos sistemas como la viga empotrada en un extremo y libre en otro en vibración transversal (Figura 3), involucran términos de elasticidad y por ello tienen un número infinito de grados de libertad.
> + La curva de deflexión elástica que ejecuta la viga la componen un número infinito de puntos de masa (Figura 4), todos los cuales realizan desplazamientos transversales (denotados por $\text{w}(x,t)$), y que se miden desde un eje neutral en equilibrio.
> + Debido a este número infinito de puntos de masa, llamémoslos: $x_{1},$ $x_{2},$ $x_{3},\ldots,$ necesitaríamos definir un número infinito de coordenadas independientes para realizar la curva de deflexión elástica.
> + De este modo, la **viga tiene un número infinito de grados de libertad**.

> + Pero, en la práctica muchos sistemas se describen usando un número finito de grados de libertad.
> + Los sistemas que tienen un número finito de grados de libertad reciben el nombre de sistemas discretos.
> + Aquellos con un número infinito de grados de libertad se llaman sistemas continuos.
> + En la práctica, los sistemas continuos se aproximan como discretos
sistemas y las soluciones se obtienen de manera simple. 
> + A pesar de que el análisis de un sistema continuo da resultados exactos, los métodos analíticos usados para tratar con sistemas continuos, sólo se limitan a una
cantidad pequeña de situaciones, como vigas, barras y placas.
> + Es por esto que muchos sistemas se analizan tratándolos como finitas masas, resortes y amortiguadores. 
> + En general, la precisión de los resultados del análisis aumenta cuando se incrementa el número de masas, resortes y amortiguadores, es decir, cuando aumenta el número de grados de libertad.

### Vector modo de vibrar:
> + Además del concepto de grados de libertad de un sistema, un segundo concepto que aparece es el de **modo de vibrar.**
> + Este concepto no está presente en sistemas de un grado de libertad.
> + Un vector modo de vibrar describe el movimiento relativo entre los grados de libertad (coordenadas).
> + Para cada frecuencia natural hay un vector modo de vibrar (Figura 5).

> + Escribir las ecuaciones en forma matricial:
$$
\ldots
$$

> + Lo cual se puede re-escribir como:

$$
\begin{align}
\text{M}{\ddot{{\bf{x}}}} + \text{K}{\bf{x}} &={\bf{0}}
\end{align}
$$

> + donde ${\ddot{{\bf{x}}}}=\left[ \ddot{x}_{1}\,\,\ddot{x}_{2}\right]^{\text{T}}$; ${\bf{x}}(t)=\left[ x_{1}\,\,x_{2}\right]^{\text{T}}$ y ${\bf{0}}=\left[ 0\,\,0\right]^{\text{T}}.$ Las matrices $\text{M}$ y $\text{K}$ se denominan matrices de masa y de rigidez respectivamente.

> + Las matrices $\text{M}$ y $\text{K}$ son simétricas, es decir, $\text{M}=\text{M}^{\text{T}}$ y $\text{K}=\text{K}^{\text{T}}$ 

> + Asumir una solución armónica: 

$$
\begin{align}
{\bf{x}}(t) & = {\bf{u}}e^{j\omega t}
\end{align}
$$

> + donde ${\bf{u}}$ es un vector de desplazamiento: ${\bf{u}} = \left[ u_{1}\,\, u_{2}\right]^{T}$ que se debe determinar y además, debe ser distinto de cero.
> + $\omega$ también debe ser determinada.
> + Derivando la Ec. (2) y reemplazando en Ec. (1) se obtiene:

$$
\begin{align*}
\left( -\omega^{2}\text{M} + \text{K}\right)\cdot{\bf{u}}e^{j\omega t} &= {\bf{0}}
\end{align*}
$$
> + Omitimos la exponencial compleja $e^{j\omega t}$ porque siempre es distinto de cero para todo $t.$
> + El vector de desplazamiento ${\bf{u}}$ también debe ser distinto de cero.
> + Para que el sistema de ecuaciones tenga solución no trivial, la matriz $\left( -\omega^{2}\text{M} + \text{K}\right)$ no debe tener inversa, porque si la tuviese, se podría pre-multiplicar por $\left( -\omega^{2}\text{M} + \text{K}\right)^{-1}$ y se obtendría ${\bf{u}}=0.$
> + La condición que la matriz $\left( -\omega^{2}\text{M} + \text{K}\right)$ no tenga inversa es equivalente a expresar que su determinante sea cero:

$$
\text{det}\,\begin{bmatrix}
-\omega^{2}m_{1}+k_{1}+k_{2} & -k_{2}\\
-k_{2} & -\omega^{2}m_{2}k_{2}
\end{bmatrix}=0
$$

> + de donde se obtiene que $\omega^{2}$ debe satisfacer:

$$
\begin{align*}
m_{1}m_{2}\omega^{4}-(m_{1}k_{2}+m_{2}k_{1}+m_{2}k_{2})+k_{1}k_{2}=0
\end{align*}
$$

> + Esta expresión se llama ecuación característica del sistema.

> + Para cada valor de $\omega^{2}$ (es decir, $\omega_{1}^{2}$ y $\omega_{2}^{2}$), se escribe un vector ${\bf{u}}$ que satisface:

$$
\begin{align*}
\left( -\omega_{1}^{2}\text{M} + \text{K}\right)\cdot{\bf{u}}_{1}&= {\bf{0}}
\end{align*}
$$
$$
\begin{align*}
\left( -\omega_{2}^{2}\text{M} + \text{K}\right)\cdot{\bf{u}}_{2}&= {\bf{0}}
\end{align*}
$$

> + Con estas expresiones obtenemos la dirección de los vectores ${\bf{u}}_{1}$ y ${\bf{u}}_{2}$ pero no su magnitud por un tema de escalamiento. Es decir, si ${\bf{u}}_{1}$ es una solución, entonces también lo es: $a{\bf{u}}_{1}$ donde $a$ es una constante distinta de cero.

> + Por lo tanto, los vectores ${\bf{u}}_{1}$ y ${\bf{u}}_{2}$ tienen magnitud arbitraria.

> + la solución general es una combinación lineal de:

$$
\begin{align*}
{\bf{x}}(t) & = \left( ae^{-j\omega_{1} t} + be^{j\omega_{1} t} \right){\bf{u}}_{1} +  \left( ce^{-j\omega_{2} t} + de^{j\omega_{2} t} \right){\bf{u}}_{2}
\end{align*}
$$

> + donde $a$, $b$, $c$ y $d$ son constantes a determinar de las condiciones iniciales.

> + También, se puede re-escribir como:

$$
\begin{align*}
{\bf{x}}(t) & = A_{1}\sin(\omega_{1}t+\phi_{1})\, {\bf{u}}_{1}+ A_{2}\sin(\omega_{2}t+\phi_{2})\, {\bf{u}}_{2}
\end{align*}
$$

> + Aquí se revela uno de los puntos más importantes del curso:

$$
\begin{align*}
\begin{bmatrix}
x_{1}\\
x_{2}
\end{bmatrix}&= {\underbrace{\begin{bmatrix}
\begin{bmatrix}
u_{11} \\
u_{21} 
\end{bmatrix} & \begin{bmatrix}
u_{12} \\
u_{22} 
\end{bmatrix}
\end{bmatrix}}_{\text{matriz de modos}}}\cdot \begin{bmatrix}
A_{1}\sin(\omega_{1}t+\phi_{1}) \\
A_{2}\sin(\omega_{2}t+\phi_{2}) 
\end{bmatrix}
\end{align*}
$$

> + Esto quiere decir, que cada masa oscila con dos frecuencias $\omega_{1}$ y $\omega_{2}.$ Estas se ordenan siempre de menor a mayor.

> + Estas frecuencias se denominan frecuencias naturales del sistema.

> + Si $A_{2}=0$, entonces cada masa oscilará a la frecuencia natural $\omega_{1}$ y la posición relativa de las masas en todo instante, va a estar dada por el vector modo de vibrar ${\bf{u}}_{1}.$ Este vector será denominado el primer modo del sistema.

> + Análogamente, si $A_{1}=0$, las masas oscilan a la frecuencia natural $\omega_{2}$ de acuerdo al vector modo de vibrar ${\bf{u}}_{2}.$

> + Este segundo vector se llama segundo modo del sistema.

### Valores propios y frecuencias naturales:

> + Asumir que tenemos un sistema con múltiples grados de libertad: 

$$
\begin{align}
\text{M}{\ddot{{\bf{x}}}} + \text{K}{\bf{x}} &={\bf{0}}
\end{align}
$$

> + Pre-multipliquemos por $\text{M}^{-1}:$

$$
\begin{align}
{\underbrace{\text{M}^{-1}\text{M}}_{\text{I}}}\,{\ddot{{\bf{x}}}} + \text{M}^{-1}\text{K}\,{\bf{x}} &={\bf{0}}\\
\text{I}\,{\ddot{{\bf{x}}}}+\text{M}^{-1}\text{K}\,{\bf{x}} &={\bf{0}} 
\end{align}
$$

> + Si suponemos una solución de la forma ${\bf{x}}(t) = {\bf{u}}e^{j\omega t}$

$$
\begin{align}
\left( \text{M}^{-1}\text{K}\, -\omega^{2}\text{I}\right)\,{\bf{u}} &={\bf{0}}
\end{align}
$$

> + Llamemos a un escalar $\lambda = \omega^{2}\text{I}$ y a una matriz $A=\text{M}^{-1}\text{K}$, entonces obtenemos:

$$
\begin{align*}
A{\bf{u}}&=\lambda {\bf{u}}
\end{align*}
$$

> + Esto es un problema de valores propios!

> + El escalar $\lambda$ se llama valor propio y el vector ${\bf{u}}$ correspondiente se llama vector propio.

> + Luego, podemos decir que las frecuencias naturales y modos de vibrar se determinan a partir de los valores propios y vectores propios, respectivamente, de la matriz $\text{M}^{-1}\text{K}$.

> + Una propiedad importante de los vectores propios es que son **ortogonales**.

> + Esto significa que **el producto punto entre dos vectores es igual a cero**:

$$
\begin{align*}
{\bf{u}}_{1}^{T}{\bf{u}}_{2} &=0
\end{align*}
$$

> + Aunque los vectores propios tienen una magnitud arbitraria, siempre se pueden normalizar (dividir por algo!) para tener largo unitario.

> + Así, podemos determinar vectores propios **ortonormales**.

> + La norma o magnitu de un vector es: 

$$
\begin{align*}
\left\Vert{\bf{x}}\right\Vert &= \sqrt{{\bf{x}}^{T}{\bf{x}}}=\left[ \sum_{i}x_{i}^{2}\right]^{1/2}
\end{align*}
$$

> + Este hecho de obtener vectores propios ortonormales **permite desacoplar las ecuaciones de movimiento**.

> + Supongamos que analizamos el caso sin amortiguamiento. Entonces, tenemos:

$$
\begin{align*}
\left(-\omega_{i}^{2}\text{M}+\text{K}\right)\,{\bf{u}}_{i} &=0
\end{align*}
$$

> + Consideremos dos modos de vibrar arbitrarios: $r$ y $s$

$$
\begin{align*}
\text{K}{\bf{u}}_{r} &= \omega_{r}^{2}\text{M}{\bf{u}}_{r}\\
\text{K}{\bf{u}}_{s} &= \omega_{s}^{2}\text{M}{\bf{u}}_{s}
\end{align*}
$$

> + Al pre-multiplicar la primera por ${\bf{u}}_{s}^{T}$ y la segunda por ${\bf{u}}_{r}^{T}$:

$$
\begin{align*}
{\bf{u}}_{s}^{T}\text{K}{\bf{u}}_{r} &= \omega_{r}^{2}{\bf{u}}_{s}^{T}\text{M}{\bf{u}}_{r}\\
{\bf{u}}_{r}^{T}\text{K}{\bf{u}}_{s} &= \omega_{s}^{2}{\bf{u}}_{r}^{T}\text{M}{\bf{u}}_{s}
\end{align*}
$$

> + La transpuesta del producto de matrices es $(A\cdot B)^{T} = B^{T}\cdot A^{T}$.

> + La transpuesta de la transpuesta de $A$ es $A$: $(A^{T})^{T} = A$.

> + La transpuesta del producto de un escalar $\alpha$ por una matriz $A$ es $(\alpha \cdot A)^{T} = \alpha\cdot A^{T}$

> + Si se transpone la segunda ecuación y se resta con la primera, obtenemos:

$$
\begin{align*}
\left( {\bf{u}}_{r}^{T}\text{K}{\bf{u}}_{s}\right)^{T} &= \left(\omega_{s}^{2}{\bf{u}}_{r}^{T}\text{M}{\bf{u}}_{s}\right)^{T}
\end{align*}
$$

$$
\begin{align*}
{\bf{u}}_{s}^{T}\text{K}{\bf{u}}_{r} &= \omega_{s}^{2}{\bf{u}}_{s}^{T}\text{M}{\bf{u}}_{r}
\end{align*}
$$

> + Al restar:

$$
\begin{align*}
0 &= \left( \omega_{r}^{2} - \omega_{s}^{2}\right) {\bf{u}}_{s}^{T}\text{M}{\bf{u}}_{r}
\end{align*}
$$

> + Esta ecuación nos da dos posibles resultados diferentes:

$$
\begin{align*}
{\bf{u}}_{s}^{T}\text{M}{\bf{u}}_{r} &= 0\quad r\neq s
\end{align*}
$$
$$
\begin{align*}
{\bf{u}}_{s}^{T}\text{M}{\bf{u}}_{r} \neq 0\quad r= s
\end{align*}
$$

> + La primera ecuación dice que dos modos distintos son ortogonales con la matriz
de masa. 

> + Y que cuando se calcula el producto del mismo modo con la matriz de masa como en la segunda ecuación, el producto no es cero. 

> + El valor de este producto se denomina **masa generalizada** $m_{ii}$:

$$
\begin{align*}
m_{ii}&={\bf{u}}_{i}^{T}\text{M}{\bf{u}}_{i}
\end{align*}
$$

> + De manera similar se define la **rigidez generalizada** $k_{ii}$:

$$
\begin{align*}
k_{ii}&={\bf{u}}_{i}^{T}\text{K}{\bf{u}}_{i}
\end{align*}
$$

> + Dado esto, obtenemos una relación importante:

$$
\begin{align*}
P^{T}MP&={\bf{m}}\\
P^{T}KP&={\bf{k}}\\
\end{align*}
$$

> + donde $P = \left[ {\bf{u}}_{1}\,\, {\bf{u}}_{2}\,\,\ldots {\bf{u}}_{n} \right]$ se denomina matriz de modos normales, ${\bf{m}}$ y ${\bf{k}}$ son matrices diagonales llamadas **matriz de masa modal** y **matriz de rigidez modal**, respectivamente.
