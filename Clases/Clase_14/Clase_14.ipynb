### Modelo del absortor amortiguado de vibraciones
> + Sugerencia de referencia [El absortor dinámico amortiguado de vibraciones](https://www.brown.edu/Departments/Engineering/Courses/En4/Notes/vibrations_mdof/vibrations_mdof.htm).
> + Recordemos que el absorbedor (o absortor) de vibraciones es un método de **control de vibraciones**.
> + Consiste de una masa y un resorte que se agregan a un **sistema principal** masa resorte de un grado de libertad.
> + Se usa para eliminar la vibración no deseada del sistema principal cuando éste se excita por una fuerza armónica.
> + Si agregamos el absorbedor de vibraciones al sistema principal de un grado de libertad, lo que resulta es un sistema de vibraciones de dos grados de libertad.
> + Por lo tanto, el sistema ahora, principal más absorbedor, tiene **dos frecuencias naturales**.
> + Si la frecuencia $\omega$ de la fuerza excitadora armónica coincide con una de las dos frecuencias naturales del sistema, éste va a entrar en **resonancia**.
> + Para superar este problema, al absorbedor dinámico se le agrega un amortiguador $c_{2}$, lo que lo convierte en un absorbedor dinámico amortiguado de vibraciones.
> + De: https://scholar.cu.edu.eg/?q=anis/files/2014-12-20_-_dynamic_absorber.pdf, la figura 1 muestra un sistema principal unido por un absorbedor dinámico amortiguado de vibraciones:

> + Usando la segunda ley de Newton para cada una de las masas: $m_{1}$ y $m_{2}$, masa principal y masa del absorbedor, respectivamente, obtenemos las ecuaciones de movimiento:

$$
\begin{align*}
m_{1} \ddot x_{1} + c_{2} \dot x_{1} - c_{2} \dot x_{2} + (k_{1}+ k_{2}) x_{1} - k_{2} x_{2} & = f_{1}(t)\\
m_{2} \ddot x_{2} - c_{2} \dot x_{1} + c_{2} \dot x_{2} - k_{2} x_{1} + k_{2} x_{2} & = 0
\end{align*}
$$

### Ecuaciones en forma matricial

$$
{\underbrace{\begin{bmatrix}
m_{1} & 0\\
0 & m_{2}
\end{bmatrix}}_{\text{matriz de masa}}}
\cdot{\overbrace{\begin{bmatrix}
\ddot x_{1} \\
\ddot x_{2}
\end{bmatrix}}^{\text{vector de aceleración}}}
+ {\underbrace{\begin{bmatrix}
c_{2} & -c_{2}\\
-c_{2} & c_{2}
\end{bmatrix}}_{\text{matriz de amortiguamiento}}}\cdot{\overbrace{\begin{bmatrix}\dot x_{1} \\
\dot x_{2}
\end{bmatrix}}^{\text{vector de velocidad}}}
+ {\underbrace{\begin{bmatrix}
k_{1} + k_{2} & -k_{2}\\
-k_{2} & k_{2}
\end{bmatrix}}_{\text{matriz de rigidez}}}\cdot{\overbrace{\begin{bmatrix}x_{1} \\
x_{2}
\end{bmatrix}}^{\text{vector de desplazamiento}}}={\overbrace{\begin{bmatrix}f_{1}(t) \\
0
\end{bmatrix}}^{\text{vector de fuerza}}}
$$

### Usando la transformada de Fourier en la representación matricial y aplicando álgebra:

$$
\begin{bmatrix}
k_{1} + k_{2} - m_{1}\omega^{2} + j\omega c_{2} & -k_{2} - j\omega c_{2}\\
-k_{2} - j\omega c_{2} & k_{2} - m_{2}\omega^{2}  + j\omega c_{2}
\end{bmatrix}\cdot\begin{bmatrix}X_{1}(\omega) \\
X_{2}(\omega)
\end{bmatrix}=\begin{bmatrix}F_{1}(\omega) \\
0
\end{bmatrix}
$$

### Despejando $X_{1}(\omega)$

> + Seguiremos la referencia: https://scholar.cu.edu.eg/?q=anis/files/2014-12-20_-_dynamic_absorber.pdf

$$
\begin{align*}
X_{1}(\omega) & = \frac{F_{1}(\omega)\cdot\left[ k_{2} - m_{2}\omega^{2} + j\omega c_{2}\right]}{\left[ k_{1} + k_{2} - m_{1}\omega^{2} + j\omega c_{2} \right]\cdot\left[ k_{2} - m_{2}\omega^{2}  + j\omega c_{2} \right] -  \left[ k_{2} + j\omega c_{2} \right]^{2}}
\end{align*}
$$

> + El denominador ($\text{den}$) lo vamos a re-organizar:

$$
\begin{align*}
{\text{den}} & = \left[\left( k_{1}-m_{1}\omega^{2} \right) + \left( k_{2}+j\omega c_{2}\right)\right]\cdot\left[\left( k_{2} - m_{2}\omega^{2}\right)  + j\omega c_{2} \right]-\left( k_{2} + j\omega c_{2} \right)^{2}\\
&= \left( k_{1}-m_{1}\omega^{2} \right)\cdot\left( k_{2} - m_{2}\omega^{2}\right)+\left( k_{1}-m_{1}\omega^{2} \right)\cdot j\omega c_{2}+\left( k_{2}+j\omega c_{2}\right)\cdot\left( k_{2} - m_{2}\omega^{2}\right)+ \left( k_{2}+j\omega c_{2}\right)\cdot j\omega c_{2}-\left( k_{2} + j\omega c_{2} \right)^{2}\\
&=\left( k_{1}-m_{1}\omega^{2} \right)\cdot\left( k_{2} - m_{2}\omega^{2}\right)+j\omega c_{2}k_{1}-j\omega^{3}m_{1}c_{2}+k_{2}^2-k_{2}m_{2}\omega^{2}+j\omega c_{2}k_{2}-j\omega^{3}m_{2}c_{2}+j\omega c_{2}k_{2}-\omega^{2}c_{2}^{2}-k_{2}^{2}-2j\omega c_{2}k_{2}+\omega^{2}c_{2}^{2}\\
&= \left[\left( k_{1}-m_{1}\omega^{2} \right)\cdot\left( k_{2} - m_{2}\omega^{2}\right)-k_{2}m_{2}\omega^{2}\right]+j\omega c_{2}\left( k_{1} - m_{1}\omega^{2} - m_{2}\omega^{2}\right)
\end{align*}
$$

> + Por lo que re-escribimos $X_{1}(\omega)$ como:

$$
\begin{align*}
X_{1}(\omega) & = \frac{F_{1}(\omega)\cdot\left[ k_{2} - m_{2}\omega^{2} + j\omega c_{2}\right]}{ \left[\left( k_{1}-m_{1}\omega^{2} \right)\cdot\left( k_{2} - m_{2}\omega^{2}\right)-k_{2}m_{2}\omega^{2}\right]+j\omega c_{2}\left( k_{1} - m_{1}\omega^{2} - m_{2}\omega^{2}\right)}
\end{align*}
$$

### Magnitud cuadrática $\left| \frac{X_{1}(\omega)}{F_{1}(\omega)}\right|^{2}:$
> + Ahora, nos basaremos en el trabajo de: http://flash.lakeheadu.ca/~kliu/JSV_2005_1.pdf.
    > + Para obtener la forma de la amplitud de respuesta normalizada, comenzaremos previamente organizando el numerado y denominador de la magnitud cuadrática.

$$
\begin{align*}
\left| \frac{X_{1}(\omega)}{F_{1}(\omega)}\right|^{2} & = \left| \frac{X_{1}(\omega)}{F_{1}(\omega)}\right|\cdot \left| \frac{X_{1}(\omega)}{F_{1}(\omega)}\right|^{*}\\
&=\frac{\left[ k_{2} - m_{2}\omega^{2} + j\omega c_{2}\right]\cdot\left[ k_{2} - m_{2}\omega^{2} - j\omega c_{2}\right]}{\left[\left[\left( k_{1}-m_{1}\omega^{2} \right)\cdot\left( k_{2} - m_{2}\omega^{2}\right)-k_{2}m_{2}\omega^{2}\right]+j\omega c_{2}\left( k_{1} - m_{1}\omega^{2} - m_{2}\omega^{2}\right)\right]\cdot\left[\left[\left( k_{1}-m_{1}\omega^{2} \right)\cdot\left( k_{2} - m_{2}\omega^{2}\right)-k_{2}m_{2}\omega^{2}\right]-j\omega c_{2}\left( k_{1} - m_{1}\omega^{2} - m_{2}\omega^{2}\right)\right]}\\
&= \frac{\left(  k_{2} - m_{2}\omega^{2} \right)^{2}+ \left( \omega c_{2} \right)^{2}}{\left[\left( k_{1}-m_{1}\omega^{2} \right)\cdot\left( k_{2} - m_{2}\omega^{2}\right)-k_{2}m_{2}\omega^{2}\right]^{2}+ \left( \omega c_{2} \right)^{2}\cdot\left( \left(k_{1} - m_{1}\omega^{2}\right) - m_{2}\omega^{2}\right)^{2}}
\end{align*}
$$

### Reducir la magnitud cuadrática a una expresión compacta adimensional:
> + Lo que haremos será en forma separada ordenar el numerado y el denominador.

### Organizamos el numerador ($\text{num}$):
> + Re-ordenamos elementos y damos vuelta los términos en el segundo paréntesis:

$$
\begin{align*}
\text{num} & = \left( \omega c_{2} \right)^{2} + \left(  k_{2} - m_{2}\omega^{2} \right)^{2}
\end{align*}
$$

> + Factorizamos cada elemento del numerador por $m_{2}\omega_{1}^{2}$

$$
\begin{align*}
\text{num} & = \left( \omega c_{2} \right)^{2} + \left(  k_{2} - m_{2}\omega^{2}\right)^{2}\\
           & = \left( \left( m_{2}\omega_{1}^{2} \right)\cdot\left( \frac{\omega c_{2}}{m_{2}\omega_{1}^{2} } \right) \right)^{2} + \left(\left( m_{2}\omega_{1}^{2} \right)\left(\frac{k_{2}}{m_{2}\omega_{1}^{2}}- \frac{\omega^{2}}{\omega_{1}^{2}} \right)\right)^{2}\\
           & = m_{2}^{2}\omega_{1}^{4}\cdot\left( \frac{\omega c_{2}}{m_{2}\omega_{1}^{2} } \right)^{2} + m_{2}^{2}\omega_{1}^{4}\cdot\left( \frac{k_{2}}{m_{2}\omega_{1}^{2}}-\frac{\omega^{2}}{\omega_{1}^{2}}\right)^{2}
\end{align*}
$$

> + Usamos $\zeta = \frac{c_{2}}{2m_{2}\omega_{1}}.$ Entonces, despejamos $\frac{c_{2}}{m_{2}\omega_{1}}=2\zeta$:
$$
\begin{align*}
\text{num} & = m_{2}^{2}\omega_{1}^{4}\cdot\left( 2\zeta\frac{\omega}{\omega_{1}} \right)^{2} + m_{2}^{2}\omega_{1}^{4}\cdot\left( \frac{k_{2}}{m_{2}\omega_{1}^{2}} - \frac{\omega^{2}}{\omega_{1}^{2}}\right)^{2}\\
           & = m_{2}^{2}\omega_{1}^{4}\cdot\left[ \left( 2\zeta\frac{\omega}{\omega_{1}} \right)^{2} + \left(\frac{\omega_{2}^{2}}{\omega_{1}^{2}} - \frac{\omega^{2}}{\omega_{1}^{2}}\right)^{2}\right]
\end{align*}
$$

> + Aquí hemos usado el hecho que: $\omega_{2}^{2} = \frac{k_{2}}{m_{2}}$

> + Sustituyendo $\beta=\frac{\omega_{2}}{\omega_{1}}$ y $r=\frac{\omega}{\omega_{1}}$, nos queda:

$$
\begin{align*}
\text{num} & = m_{2}^{2}\omega_{1}^{4}\cdot\left[ \left( 2\zeta r \right)^{2} + \left( \beta^{2}- r^{2}\right)^{2}\right]
\end{align*}
$$

### Organizamos el denominador ($\text{den}$):

> + $\text{den} = \text{den}_{1} + \text{den}_{2}$

### Comenzamos organizando $\text{den}_{1}$:
$$
\begin{align*}
\text{den}_{1} = \left( \omega c_{2} \right)^{2}\cdot\left( \left( k_{1} - m_{1}\omega^{2}\right) - m_{2}\omega^{2}\right)^{2}
\end{align*}
$$

> + Factorizamos por $k_{1}$ el segundo paréntesis:

$$
\begin{align*}
\text{den}_{1} & = \left( \omega c_{2} \right)^{2}\cdot\left( k_{1}\left( \left(1- \frac{m_{1}}{k_{1}}\omega^{2}\right) - \frac{m_{2}}{k_{1}}\omega^{2} \right) \right)^{2}\\
               & = k_{1}^{2}\cdot \left( \omega c_{2} \right)^{2}\cdot\left( \left(1- \frac{m_{1}}{k_{1}}\omega^{2}\right) -  \frac{m_{2}}{k_{1}}\omega^{2}\right)^{2}
\end{align*}
$$

> + Sustituyendo $\mu = \frac{m_{2}}{m_{1}}$, entonces, $m_{2}= \mu m_{1}=\mu \frac{k_{1}}{\omega_{1}^{2}}.$ Por lo que, $\frac{m_{2}}{k_{1}}=\frac{\mu}{\omega_{1}^{2}}$

> + Usaremos también el hecho que: $\omega_{1}^{2} = \frac{k_{1}}{m_{1}}$ y la relación $r=\frac{\omega}{\omega_{1}}$. Entonces, nos queda:

$$
\begin{align*}
\text{den}_{1} & = k_{1}^{2}\cdot \left( \omega c_{2} \right)^{2}\cdot\left( \left(1- r^{2}\right)- \mu r^{2}\right)^{2} \\
               & = k_{1}^{2}\cdot \left( \omega c_{2} \right)^{2}\cdot\left( 1- r^{2} - \mu r^{2}\right)^{2}\\
               & = k_{1}^{2}\cdot \left( \omega c_{2} \right)^{2}\cdot\left( 1- \left( 1+\mu\right)r^{2} \right)^{2}
\end{align*}
$$

> + Arreglamos ahora, el témino $\left( \omega c_{2} \right)^{2}.$ Usamos el hecho que $c_{2} = 2\zeta m_{2}\omega_{1}$ 
$$
\begin{align*}
\left( \omega c_{2} \right)^{2} & = \left( \omega\cdot 2\zeta m_{2}\omega_{1} \right)^{2}\\
                                & = \left( m_{2}\omega_{1} \left(2\zeta\omega\right)\right)^{2},\quad {\text{Truco: multiplicamos y dividimos por} \,\,\omega_{1}}\\
                                & = \left( m_{2}\omega_{1} \left(2\zeta\omega\right)\frac{\omega_{1}}{\omega_{1}}\right)^{2}\\
                                & = \left( m_{2}\omega_{1}^{2} \left(2\zeta\frac{\omega}{\omega_{1}}\right)\right)^{2}\\
                                & = \left( m_{2}\omega_{1}^{2} \left(2\zeta r\right)\right)^{2}
\end{align*}
$$

> + Por lo que:

$$
\begin{align*}
\left( \omega c_{2} \right)^{2} & = m_{2}^{2}\omega_{1}^{4}\cdot \left(2\zeta r\right)^{2}
\end{align*}
$$

> + Así, $\text{den}_{1}$ nos queda igual a:

$$
\begin{align*}
\text{den}_{1} & = k_{1}^{2}\cdot \left( \omega c_{2} \right)^{2}\cdot\left( 1- \left( 1+\mu\right)r^{2} \right)^{2}\\
               & = k_{1}^{2} m_{2}^{2}\omega_{1}^{4}\cdot \left(2\zeta r\right)^{2}\cdot\left( 1- \left( 1+\mu\right)r^{2} \right)^{2}
\end{align*}
$$

### Organizamos ahora $\text{den}_{2}:$

$$
\begin{align*}
\text{den}_{2} &= \left[ \left( k_{1}-m_{1}\omega^{2}  \right)\cdot \left( k_{2}-m_{2}\omega^{2} \right) - k_{2} m_{2}\omega^{2}\right]^{2}
\end{align*}
$$

> + Factorizamos $\left(k_{1}m_{2}\omega_{1}^{2}\right)$:

$$
\begin{align*}
\text{den}_{2} &= \left[ k_{1}\left( 1 - \frac{m_{1}}{k_{1}}\omega^{2}\right)\cdot m_{2}\omega_{1}^{2}\left(\frac{k_{2}}{m_{2}\omega_{1}^{2}} - \frac{\omega^{2}}{\omega_{1}^{2}} \right) - \left(k_{1}m_{2}\omega_{1}^{2}\right)\left(\frac{k_{2}\omega^{2}}{k_{1}\omega_{1}^{2}}\right)\right]^{2}\\
& = \left[ k_{1}m_{2}\omega_{1}^{2}\cdot \left[ \left( 1-\frac{m_{1}}{k_{1}}\omega^{2} \right)\cdot\left( \frac{\omega_{2}^{2}}{\omega_{1}^{2}}-\frac{\omega^{2}}{\omega_{1}^{2}}\right) - \left(\frac{k_{2}\omega^{2}}{k_{1}\omega_{1}^{2}}\right)\right] \right]^{2}
\end{align*}
$$

> + Sustituimos: $\beta=\frac{\omega_{2}}{\omega_{1}}$, $r=\frac{\omega}{\omega_{1}}$ y $\frac{k_{2}}{k_{1}}=\mu\frac{\omega_{2}^{2}}{\omega_{1}^{2}}$ de tal manera que:

$$
\begin{align*}
\text{den}_{2} &=  k_{1}^{2}m_{2}^{2}\omega_{1}^{4}\cdot \left[ \left( 1-r^{2} \right)\cdot\left( \beta^{2}-r^{2}\right) - \mu\beta^{2}r^{2} \right]^{2}
\end{align*}
$$

### Juntamos los términos del denominador:

$$
\begin{align*}
\text{den} & = \text{den}_{1} + \text{den}_{2}\\
           & = k_{1}^{2} m_{2}^{2}\omega_{1}^{4}\cdot\left[ \left(2\zeta r\right)^{2}\cdot\left( 1- \left( 1+\mu\right)r^{2} \right)^{2}+\left[ \left( 1-r^{2} \right)\cdot\left( \beta^{2}-r^{2}\right) - \mu\beta^{2}r^{2} \right]^{2}\right]\\
           & = k_{1}^{2} m_{2}^{2}\omega_{1}^{4}\cdot\left(\left[ 1- \left( 1+\mu\right)r^{2} \right]^{2}\left(2\zeta r\right)^{2}+\left[ \left( 1-r^{2} \right)\cdot\left( \beta^{2}-r^{2}\right) - \mu\beta^{2}r^{2} \right]^{2}\right)
\end{align*}
$$

### Re-escribimos la magnitud cuadrática:

$$
\begin{align*}
\left| \frac{X_{1}(\omega)}{F_{1}(\omega)}\right|^{2} & = \frac{\left(  k_{2} - m_{2}\omega^{2} \right)^{2}+ \left( \omega c_{2} \right)^{2}}{\left[\left( k_{1}-m_{1}\omega^{2} \right)\cdot\left( k_{2} - m_{2}\omega^{2}\right)-k_{2}m_{2}\omega^{2}\right]^{2}+ \left( \omega c_{2} \right)^{2}\cdot\left( \left(k_{1} - m_{1}\omega^{2}\right) - m_{2}\omega^{2}\right)^{2}}\\
& = \frac{m_{2}^{2}\omega_{1}^{4}\cdot\left[ \left( 2\zeta r \right)^{2} + \left( \beta^{2}- r^{2}\right)^{2}\right]}{k_{1}^{2} m_{2}^{2}\omega_{1}^{4}\cdot\left(\left[ 1- \left( 1+\mu\right)r^{2} \right]^{2}\left(2\zeta r\right)^{2}+\left[ \left( 1-r^{2} \right)\cdot\left( \beta^{2}-r^{2}\right) - \mu\beta^{2}r^{2} \right]^{2}\right)}\\
& = \frac{ \left( 2\zeta r \right)^{2} + \left( \beta^{2}- r^{2}\right)^{2}}{k_{1}^{2} \cdot\left(\left[ 1- \left( 1+\mu\right)r^{2} \right]^{2}\left(2\zeta r\right)^{2}+\left[ \left( 1-r^{2} \right)\cdot\left( \beta^{2}-r^{2}\right) - \mu\beta^{2}r^{2} \right]^{2}\right)}
\end{align*}
$$

### Normalizamos $F_{1}(\omega)$ por $k_{1}$, es decir, lo dividimos por $k_{1}$ y expresamos la magnitud:

$$
\begin{align*}
\left| \frac{X_{1}(\omega)}{\frac{F_{1}(\omega)}{k_{1}}}\right| & = \sqrt{\frac{ \left( 2\zeta r \right)^{2} + \left( \beta^{2}- r^{2}\right)^{2}}{\left[ 1- \left( 1+\mu\right)r^{2} \right]^{2}\left(2\zeta r\right)^{2}+\left[ \left( 1-r^{2} \right)\cdot\left( \beta^{2}-r^{2}\right) - \mu\beta^{2}r^{2} \right]^{2}}}
\end{align*}
$$
